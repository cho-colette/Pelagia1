```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

mean_cm = pd.read_csv("/content/Pelagia1/results/stats/mean_confusion_matrix.csv", index_col=0)
out_dir = Path("/content/Pelagia1/results/stats")

fig, ax = plt.subplots(figsize=(9, 7))
im = ax.imshow(mean_cm.values)

ax.set_xticks(range(len(mean_cm.columns)))
ax.set_yticks(range(len(mean_cm.index)))
ax.set_xticklabels([str(c).replace("_", " ") for c in mean_cm.columns], rotation=30, ha="right")
ax.set_yticklabels([str(i).replace("_", " ") for i in mean_cm.index])

ax.set_xlabel("Predicted label")
ax.set_ylabel("True label")
ax.set_title("Mean confusion matrix across repeated stochastic runs")

for i in range(mean_cm.shape[0]):
    for j in range(mean_cm.shape[1]):
        ax.text(j, i, f"{mean_cm.iloc[i, j]:.2f}", ha="center", va="center")

fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
fig.tight_layout()
fig.savefig(out_dir / "mean_confusion_matrix.png", dpi=300, bbox_inches="tight")
plt.show()
```
